```python
import matplotlib.pyplot as plt
import numpy as np
```

# _Task 1_

Plot the populations for the years in the period. Add legend and axis labels to the plot.

```python
data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')

years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]
```

```python
plt.figure(figsize=(10,5))


plt.plot(years, hares, label='Hare')
plt.plot(years, lynxes, label='Lynx')
plt.plot(years, carrots, label='Carrot')

plt.legend()
plt.xticks(np.arange(years.astype(int)[0],years.astype(int)[-1]+1, 2))

plt.show()
```

# _Task 2_

Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for the years in the period.

```python
plt.figure(figsize=(10,5))

norm = hares + lynxes
plt.bar(years.astype(int), hares/norm, label='hare')
plt.bar(years.astype(int), lynxes/norm, bottom = hares/norm, label='lynx', color = '#6EC06E')
plt.yticks([0.0, 0.25, 0.5, 0.75, 1.0])
plt.ylabel('Ratio', fontsize=20)
plt.xlabel('Years', fontsize=20)
plt.legend(shadow = False, fontsize=15)

plt.show()
```

# _Task 3_


Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

```python
print('Mean of hares: ', np.mean(hares), '\nStd of hares:  ', np.std(hares), '\n')
print('Mean of lynxes: ', np.mean(lynxes), '\nStd of lynxes:  ', np.std(lynxes), '\n')
print('Mean of carrots: ', np.mean(carrots), '\nStd of carrots:  ', np.std(carrots), '\n')
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.barh(years.astype(int), hares, label='Hares')
plt.vlines(np.mean(hares),years.astype(int)[0],years.astype(int)[-1], color='r')
plt.legend()
plt.ylabel('Years')
plt.xlabel('Hares')

plt.subplot(1,3,2)
plt.barh(years.astype(int), lynxes, label='Lynxes', color = '#6EC06E')
plt.vlines(np.mean(lynxes),years.astype(int)[0],years.astype(int)[-1], color='r')
plt.legend()
plt.xlabel('Lynxes')

plt.subplot(1,3,3)
plt.barh(years.astype(int), carrots, label='Carrots')
plt.vlines(np.mean(carrots),years.astype(int)[0],years.astype(int)[-1], color='r')
plt.legend()
plt.xlabel('Carrots')

plt.show()
```

# _Task 4_

Find which year each species had the largest population. Mark these years in the plot of populations.

```python
plt.figure(figsize=(10,5))

hare_max = np.argmax(hares)
lynx_max = np.argmax(lynxes)
carrot_max = np.argmax(carrots)


plt.plot(years, hares, label='Hare')
plt.plot(years, lynxes, label='Lynx')
plt.plot(years, carrots, label='Carrot')
plt.scatter([years[hare_max], years[lynx_max], years[carrot_max]],
            [hares[hare_max], lynxes[lynx_max], carrots[carrot_max]], color='r')

plt.legend()
plt.xticks(np.arange(years.astype(int)[0],years.astype(int)[-1]+1, 2))

plt.show()
```

Plot the change in hare population and the number of lynxes. Find the correlation coefficient.

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
images = digits.data.reshape(-1, 8, 8)
```

```python
i = 985
plt.imshow(images[i], cmap='gray')
plt.title('This is {}'.format(digits.target[i]))
plt.show()
```

```python
from sklearn.decomposition import PCA
n_components=20
pca = PCA(n_components)

X_transformed = pca.fit_transform(digits.data.astype(float))
X_transformed.shape
```

```python
plt.figure(figsize = (15, 5))
plt.bar(np.arange(n_components), pca.explained_variance_)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=digits.target.astype(int), 
            cmap=plt.cm.get_cmap('hsv', 10))
plt.colorbar()
plt.show()
```

```python
pca.components_.shape, pca.mean_.shape
```

```python
X_restored = pca.inverse_transform(X_transformed)
X_restored.shape
```

```python
i = 540
fig, ax = plt.subplots(1, 2)
ax[0].imshow(digits.images[i], cmap='gray')
ax[1].imshow(X_restored[i].reshape(8, 8), cmap='gray')
ax[0].set_title('Original')
ax[1].set_title('Restored')
plt.show()
```

```python
var = []
rng = np.arange(10, 65, 5)
for components in rng:
    pca = PCA(components)
    X_transformed = pca.fit_transform(digits.data.astype(float)) 
    var.append(pca.explained_variance_ratio_.sum())
var = np.asarray(var) * 100
plt.plot(rng, var, '-o')
plt.xlabel('n_components')
plt.ylabel('total explained variance')
plt.show()
```

```python
from sklearn.datasets import fetch_openml
digits = fetch_openml('mnist_784')
```

```python
rng = [16, 20, 27, 45]

for i in rng:
    pca = PCA(n_components=i)
    X_transformed = pca.fit_transform(digits.data.astype(float))
    X_restored = pca.inverse_transform(X_transformed[0])
    fig, ax = plt.subplots(1, 2)
    ax[0].imshow(digits.data.astype(float)[0].reshape(8, 8), cmap='gray')
    ax[0].set_title('Original')
    ax[1].imshow(X_restored.reshape(8, 8), cmap='gray')
    ax[1].set_title('Restored')
    
    plt.show()
```
